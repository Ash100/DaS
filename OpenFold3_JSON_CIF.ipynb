##OpenFold3 .JSON to .CIF

This notebook is generated by **Dr. Ashfaq Ahmad**. If you have .JSON file downloaded from the OpenFold3, below we can convert them into .CIF file. The .CIF file can be viewed and analyzed in Discovery visualizer, PyMol, Chimera and other visualizers. In case you are interested only in PDB format, you can save the structure into PDB format using your favourite viewer.

```python
!pip install biopython
```

```python
import json

# Load the JSON file
with open("/content/AKT-L.json") as f:
    data = json.load(f)

# Extract mmCIF string
mmcif_string = data["outputs"][0]["structures_with_scores"][0]["structure"]

# Save to .cif file
with open("AKT_structure.cif", "w") as f:
    f.write(mmcif_string)
```
